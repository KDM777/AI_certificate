```python
path='/content/drive/MyDrive/실습용자료'

df = pd.read_csv('breast_cancer.csv')
df.info()
```

```python

x = df.iloc[:,0:9]

y=df.iloc[:, 9]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test= train_test_split(x,y) # 3:1로 나누기
sns.pairplot(df, hue='class')
```

```python
from sklearn.neighbors import KNeighborsClassifier

#모델 설정
knn = KNeighborsClassifier() # knn에 모델을 저장

#모델 훈련
knn.fit(x_train, y_train)

print("train score", knn.score(x_train,y_train))
print("test score", knn.score(x_test,y_test)) # 테스트 스코어

y_pred=knn.predict(x_test)
y_pred
```

```python
# k 값은 홀수로 정한다.
k_list=list(range(3,16,2))
score_list=[]

for k in k_list:
  knn = KNeighborsClassifier(n_neighbors=k) # knn에 모델을 저장
  knn.fit(x_train,y_train)
  score_list.append(knn.score(x_test,y_test))


```

```python
plt.plot(k_list,score_list)
```

```python

#모델 설정
knn = KNeighborsClassifier(n_neighbors=3) # knn에 모델을 저장

#모델 훈련
knn.fit(x_train, y_train)

print("train score", knn.score(x_train,y_train))
print("test score", knn.score(x_test,y_test)) # 테스트 스코어
```

```python
import joblib
model=knn.fit(x_train, y_train)



joblib.dump(model, "KNN_model.pkl") # 현재 작업하는 파일에 저장
```

```python
df = pd.read_csv('iris.csv')
```

```python
x=df.iloc[:, :4]
y=df.iloc[:, 4]

x_train, x_test, y_train, y_test=train_test_split(x,y,
                                                  random_state=10,  # 랜덤 추출해도 똑같이
                                                  test_size=0.2
                                                  )
```

```python
from sklearn.tree import DecisionTreeClassifier

tree_clf=DecisionTreeClassifier()
tree_clf.fit(x_train,y_train)

print("train score", tree_clf.score(x_train,y_train))
print("test score", round(tree_clf.score(x_test,y_test)*100,2), "%")
```

```python
print(tree_clf.feature_names_in_)
print(tree_clf.feature_importances_)
```

```python
import matplotlib.pyplot as plt
import numpy as np
n_features = x_train.shape[1]

plt.barh(np.arange(n_features), tree_clf.feature_importances_, align="center")
plt.yticks(np.arange(n_features),tree_clf.feature_names_in_);
```

```python
class_names=df['Species'].unique()
```

```python
import graphviz
from sklearn.tree import export_graphviz

dot_data = export_graphviz(
    tree_clf,
    feature_names = tree_clf.feature_names_in_,
    class_names = class_names,
    rounded = True,
    filled=True,
    proportion= True
)

display(graphviz.Source(dot_data))
```

```python
tree_clf=DecisionTreeClassifier(max_depth=3)
tree_clf.fit(x_train,y_train)

print("train score", round(tree_clf.score(x_train,y_train),3 ))
print("test score", round(tree_clf.score(x_test,y_test),3))
```

```python
import graphviz
from sklearn.tree import export_graphviz

dot_data = export_graphviz(
    tree_clf,
    feature_names = tree_clf.feature_names_in_,
    class_names = class_names,
    rounded = True,
    filled=True,
    proportion= True
)

display(graphviz.Source(dot_data))
```
